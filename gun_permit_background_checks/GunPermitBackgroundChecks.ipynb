```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('nics-firearm-background-checks.csv')
```

```python
df
```

```python
for column in df.columns:
  print(column)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.corr(numeric_only=True)
```

```python
x_arr = np.array(df['handgun'])
y_arr = np.array(df['long_gun'])
x_label = 'handgun'
y_label = 'long_gun'
plt.xlabel(x_label)
plt.ylabel(y_label)
a, b = np.polyfit(x_arr, y_arr, deg=1)
plt.scatter(x_arr, y_arr)
plt.plot(x_arr, a*x_arr + b)
print(f'The equation of the line is y = {np.round(a, 3)}x + {np.round(b, 3)}')
```

```python
alabama_df = df[df['state'] == 'Alabama']
alabama_df.reset_index(drop=True, inplace=True)
alabama_df
```

```python
alabama_df = alabama_df[::-1].reset_index(drop=True)
alabama_df
```

```python
months = alabama_df.index[alabama_df['state'] == 'Alabama'].tolist()
alabama_df['months_since_start'] = months
alabama_df
```

```python
alabama_df = alabama_df[['months_since_start', 'month', 'state', 'permit', 'permit_recheck', 'handgun', 'long_gun',
       'other', 'multiple', 'admin', 'prepawn_handgun', 'prepawn_long_gun',
       'prepawn_other', 'redemption_handgun', 'redemption_long_gun',
       'redemption_other', 'returned_handgun', 'returned_long_gun',
       'returned_other', 'rentals_handgun', 'rentals_long_gun',
       'private_sale_handgun', 'private_sale_long_gun', 'private_sale_other',
       'return_to_seller_handgun', 'return_to_seller_long_gun',
       'return_to_seller_other', 'totals']]
alabama_df
```

```python
alabama_df.corr(numeric_only=True)
```

```python
alabama_x_arr = np.array(alabama_df['months_since_start'])
alabama_y_arr = np.array(alabama_df['multiple'])
alabama_x_label = 'Months Since Start'
alabama_y_label = 'Multiple-Gun Background Checks'
plt.xlabel(alabama_x_label)
plt.ylabel(alabama_y_label)
alabama_a, alabama_b = np.polyfit(alabama_x_arr, alabama_y_arr, 1)
plt.scatter(alabama_x_arr, alabama_y_arr)
plt.plot(alabama_x_arr, alabama_a*alabama_x_arr + alabama_b, color='r')
print(f'Correlation between Alabama\'s tracked months and multiple-gun background checks modeled with y = {np.round(alabama_a, 3)}x + {np.round(alabama_b, 3)}')
```
